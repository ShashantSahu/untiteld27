```python
import PIL.Image

# Ordered from darkest to lightest characters
ASCII_CHARS = ["@", "#", "S", "%", "?", "*", "+", ";", ":", ",", "."]

def process_james_hetfield(path, new_width=100):
    try:
        # 1. Open and grayscale (L = Luminance mode)
        img = PIL.Image.open(path).convert('L')

        # 2. Manual Resize Logic to handle aspect ratio
        width, height = img.size
        aspect_ratio = height / width
        # Multiply height by 0.5 because characters are taller than they are wide
        new_height = int(new_width * aspect_ratio * 0.5)
        img = img.resize((new_width, new_height))

        # 3. Access pixel data directly for manual processing
        pixels = img.load()
        ascii_str = ""

        # Outer Loop: Rows (y-axis)
        for y in range(new_height):
            # Inner Loop: Columns (x-axis)
            for x in range(new_width):
                pixel_value = pixels[x, y] # Value from 0-255

                # Manual Mapping Logic (Conditional Logic)
                # 255 / 25 = ~10.2 (maps 0-255 to index 0-10)
                index = pixel_value // 25

                # Bound check to avoid index errors
                if index > 10: index = 10

                ascii_str += ASCII_CHARS[index]

            ascii_str += "\n" # End of row

        return ascii_str

    except Exception as e:
        return f"Error: {e}"

# Running the code in Colab or Local
path = "/content/sample_data/James Hetfield.16.12.jpeg" # Change this to your copied cloud path
output = process_james_hetfield(path)
print(output)
```
